```python
import pandas as pd

def adjust_timestamp(ts):
    """
    Adjust timestamps with hour '24' to proper format:
    - Replace '24:' with '00:'
    - Add one day to the date portion if a date is present.
    """
    # Trim any leading/trailing whitespace
    ts = ts.strip()
    
    # Check if time part contains '24:' 
    if "24:" in ts:
        # Assume the timestamp is in the format "YYYY-MM-DD 24:MM:SS" or similar
        try:
            date_str, time_str = ts.split()
        except ValueError:
            # If ts only contains time (unlikely for full timestamps), handle accordingly
            time_str = ts
            date_str = None
        
        # Replace '24:' with '00:'
        new_time_str = time_str.replace("24:", "00:", 1)
        
        # If date information is available, convert the date, add one day, and rebuild the string
        if date_str is not None:
            try:
                date_obj = pd.to_datetime(date_str)
            except Exception as e:
                raise ValueError(f"Unable to parse date part {date_str}: {e}")
            # Add one day to the date
            date_obj += pd.Timedelta(days=1)
            # Combine adjusted date with new time
            new_ts = f"{date_obj.strftime('%Y-%m-%d')} {new_time_str}"
        else:
            new_ts = new_time_str
    else:
        new_ts = ts

    # Return the corrected datetime
    return pd.to_datetime(new_ts)

# Example DataFrame:
data = {
    ' Timestamp': ['2025-04-15 23:59:59', '2025-04-15 24:03.1', '2025-04-15 22:15:30']
}
df = pd.DataFrame(data)

# Apply conversion with adjustment
df['Timestamp'] = df[' Timestamp'].apply(adjust_timestamp)

# Extract useful time features
df['Hour'] = df['Timestamp'].dt.hour
df['Minute'] = df['Timestamp'].dt.minute
df['Second'] = df['Timestamp'].dt.second
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek

print(df)
print("Time features created")
```

```python
# Create packet ratios
df['Packet_Ratio'] = df[' Total Fwd Packets'] / (df[' Total Backward Packets'] + 1)

# Create byte ratios
df['Byte_Ratio'] = df['Total Length of Fwd Packets'] / (df[' Total Length of Bwd Packets'] + 1)

# Create packet length ratio
df['Packet_Length_Ratio'] = df[' Fwd Packet Length Mean'] / (df[' Bwd Packet Length Mean'] + 1)

# Create flag features
flag_columns = ['FIN Flag Count', ' SYN Flag Count', ' RST Flag Count', 
                ' PSH Flag Count', ' ACK Flag Count', ' URG Flag Count']
df['Total_Flags'] = df[flag_columns].sum(axis=1)

# Create IAT (Inter-Arrival Time) features
df['IAT_Ratio'] = df[' Fwd IAT Mean'] / (df[' Bwd IAT Mean'] + 1)

print(f"Feature engineering completed. Dataset now has {df.shape[1]} columns")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure column names have no extra spaces
df.columns = df.columns.str.strip()

# Group data by 'Label' and compute the sum for selected metrics.
grouped = df.groupby(' Label')[[' Total Fwd Packets', ' Total Backward Packets', 
                                ' Total Length of Fwd Packets', ' Total Length of Bwd Packets']].sum().reset_index()

# Set the color palette with bright colors.
colors = sns.color_palette("bright", 4)

# Create a stacked bar chart.
plt.figure(figsize=(12,8))
bar_width = 0.6
labels = grouped[' Label']

# Bottom for stacking: first metric.
bottom = grouped[' Total Fwd Packets']
plt.bar(labels, grouped[' Total Fwd Packets'], color=colors[0], label=' Total Fwd Packets', width=bar_width)

# Next, stack Total Backward Packets on top.
plt.bar(labels, grouped[' Total Backward Packets'], color=colors[1],
        bottom=bottom, label=' Total Backward Packets', width=bar_width)
bottom += grouped[ ' Total Backward Packets']

# Stack Total Length of Fwd Packets.
plt.bar(labels, grouped[' Total Length of Fwd Packets'], color=colors[2],
        bottom=bottom, label=' Total Length of Fwd Packets', width=bar_width)
bottom += grouped[' Total Length of Fwd Packets']

# Finally, stack Total Length of Bwd Packets.
plt.bar(labels, grouped[' Total Length of Bwd Packets'], color=colors[3],
        bottom=bottom, label=' Total Length of Bwd Packets', width=bar_width)

plt.title("Stacked Bar Chart: Aggregated Metrics by Label", fontsize=18, color="darkblue")
plt.xlabel("Label", fontsize=14)
plt.ylabel("Sum of Metrics", fontsize=14)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
